```python
import codecs
# from konlpy.tag import Twitter
from konlpy.tag import Okt
# utf-8 인코딩으로 파일을 열고 글자를 출력하기 --- (※1)
# fp = codecs.open("hong.txt", "r", encoding="utf-8")
# fp = codecs.open("./data/박근혜_국정연설문_2016.txt", "r", encoding="utf-8")
fp = codecs.open("음식.csv", "r", encoding="utf-8")
text = fp.read()
# 텍스트를 한 줄씩 처리하기 --- (※2)
# twitter = Twitter()
okt = Okt()
word_dic = {}
lines = text.split("\r\n")
```

```python
from konlpy.tag import Okt

okt = Okt()
word_dic = {}

for line in lines:
    malist = okt.pos(line, norm=True, stem=True)
    for word in malist:
        if word[1] == "Noun" and len(word[0]) >= 2:  # 두 글자 이상인 명사만 고려
            if word[0] not in word_dic:
                word_dic[word[0]] = 0
            word_dic[word[0]] += 1

# 가장 많이 사용된 명사 상위 50개만 추출
top_n = 50
keys = sorted(word_dic.items(), key=lambda x: x[1], reverse=True)[:top_n]

for word, count in keys:
    print("{0}({1})".format(word, count))
print()
```

```python
from konlpy.tag import Okt
from wordcloud import WordCloud
from collections import Counter
import matplotlib.pyplot as plt
```

```python
# Word Cloud 생성 및 시각화
font_path = 'NanumBarunGothic.ttf'
wordcloud = WordCloud(
    font_path=font_path,
    width=800,
    height=800,
    background_color="white"
)

# 수정: word_list 대신 word_dic
wordcloud.generate_from_frequencies(word_dic)

array = wordcloud.to_array()

fig = plt.figure(figsize=(10, 10))
plt.imshow(array, interpolation="bilinear")
plt.axis('off')  # 축 제거
plt.show()
fig.savefig('wordcloud.png', dpi=300)
```
