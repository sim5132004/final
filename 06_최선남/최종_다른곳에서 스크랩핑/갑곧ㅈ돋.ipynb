### Okt

```python
import codecs
from konlpy.tag import Twitter
from konlpy.tag import Okt
from gensim.models import word2vec 
```

```python
from gensim.models import word2vec 
# utf-16 인코딩으로 파일을 열고 글자를 출력하기 --- (※1)
# 텍스트를 한 줄씩 처리하기 --- (※2)
# twitter = Twitter()
okt = Okt()
results = []
lines = cleaned_text.split('\t')
for line in lines:
    # 형태소 분석하기 --- (※3)
    # 단어의 기본형 사용
    malist = okt.pos(line, norm=False, stem=False)
    r = []
    for word in malist:
        # 어미/조사/구두점 등은 대상에서 제외 
        if not word[1] in ["Josa", "Eomi", "Punctuation"]:
            r.append(word[0])
    rl = (" ".join(r)).strip()
    results.append(rl)
    print(rl)

    wakati_file = 'gab.wakati'

with open(wakati_file, 'w', encoding='utf-8') as fp:
    fp.write("\n".join(results))
# Word2Vec 모델 만들기 --- (※5)
data = word2vec.LineSentence(wakati_file)
model = word2vec.Word2Vec(data, window=10, hs=1, min_count=5, sg=1)
# 예제 옵션을 활용하며 기타 옵션은 메뉴얼 참고
# The parameters:
# min_count = int - Ignores all words with total absolute frequency lower than this - (2, 100)
# window = int - The maximum distance between the current and predicted word within a sentence. E.g. window words on the left and window words on the left of our target - (2, 10)
# size = int - Dimensionality of the feature vectors. - (50, 300)
# sample = float - The threshold for configuring which higher-frequency words are randomly downsampled. Highly influencial. - (0, 1e-5)
# alpha = float - The initial learning rate - (0.01, 0.05)
# min_alpha = float - Learning rate will linearly drop to min_alpha as training progresses. To set it: alpha - (min_alpha * epochs) ~ 0.00
# negative = int - If > 0, negative sampling will be used, the int for negative specifies how many "noise words" should be drown. If set to 0, no negative sampling is used. - (5, 20)
# workers = int - Use these many worker threads to train the model (=faster training with multicore machines)

model.save("gab.model")
# model.save("moon.model")
print("\n\n================= 분석 완료 ==================")
```

```python
import codecs
# from konlpy.tag import Twitter
from konlpy.tag import Okt
# utf-8 인코딩으로 파일을 열고 글자를 출력하기 --- (※1)
# fp = codecs.open("hong.txt", "r", encoding="utf-8")
# fp = codecs.open("./data/박근혜_국정연설문_2016.txt", "r", encoding="utf-8")
fp = codecs.open("갑곶돈.csv", "r", encoding="utf-8")
text = fp.read()
# 텍스트를 한 줄씩 처리하기 --- (※2)
# twitter = Twitter()
okt = Okt()
word_dic = {}
lines = text.split("\r\n")
```

```python
from konlpy.tag import Okt

okt = Okt()
word_dic = {}

for line in lines:
    malist = okt.pos(line, norm=True, stem=True)
    for word in malist:
        if word[1] == "Noun" and len(word[0]) >= 2:  # 두 글자 이상인 명사만 고려
            if word[0] not in word_dic:
                word_dic[word[0]] = 0
            word_dic[word[0]] += 1

# 가장 많이 사용된 명사 상위 50개만 추출
top_n = 50
keys = sorted(word_dic.items(), key=lambda x: x[1], reverse=True)[:top_n]

for word, count in keys:
    print("{0}({1})".format(word, count))
print()
```

## 로지스틱회귀

```python
import pandas as pd
import warnings
warnings.filterwarnings('ignore')
fish = pd.read_csv('갑족곧.csv')
fish.head()
```

```python
print(pd.unique(fish['키워드']))
```

```python
fish_input = fish[['횟수']].to_numpy()
```

```python
print(fish_input[:5])
```

```python
fish_target = fish['키워드'].to_numpy()
```

```python
from sklearn.model_selection import train_test_split

train_input, test_input, train_target, test_target = train_test_split(
    fish_input, fish_target, random_state=42)
```

```python
from sklearn.preprocessing import StandardScaler

ss = StandardScaler()
ss.fit(train_input)
train_scaled = ss.transform(train_input)
test_scaled = ss.transform(test_input)
```

```python
from sklearn.neighbors import KNeighborsClassifier

kn = KNeighborsClassifier(n_neighbors=3)
kn.fit(train_scaled, train_target)

print(kn.score(train_scaled, train_target))
print(kn.score(test_scaled, test_target))
```
