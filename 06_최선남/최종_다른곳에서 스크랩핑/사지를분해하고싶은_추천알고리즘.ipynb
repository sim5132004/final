## 캐글 movie 데이터셋으로 연습해보기

```python
movie_df=pd.read_csv('tmdb_5000_movies.csv')
```

```python
movie_df.head(5)
```

```python
data=movie_df[['id','genres','vote_average','vote_count','popularity','title','keywords','overview']]
```

```python
data.head(5)
```

## 전처리

* 평균평점이 불공정 처리 되어있음
* 데이터를 제공해준 imdb에서 제안한 weight rating으로 평점 다시 처리
* 그 공식 적용해서 socre컬럼 다시만듬

```python
## 500위안에 들어올려면 vote_count가 상위 몇퍼인가 
m=data['vote_count'].quantile(0.9) ##quantile(0.9) => 데이터의 분포에서 90% 지점에 해당하는 값을 의미함. 하위 90%임 
#quantile은 함수의 데이터 분위스를 계상하는데 사용된다. 0.5는 중앙값, 0.25는 1사분위를 계산함
print(m) 
```

```python
data=data.loc[data['vote_count']>=m]
```

```python
data.shape
```

```python
c=data['vote_average'].mean()
```

```python
print(c)
```

```python
def weight_rating(x, m=m,c=c):
    v=x['vote_count']
    r=x['vote_average']
    
    return(v/(v+m)*r)+(m/(v+m)*c)
```

```python
data['score']=data.apply(weight_rating, axis=1)
```

```python
data.head()
```

## 콘텐츠기반 필터링을 위한 컬럼추출

* 사용자가 특정아이템을 선택한경우, 그 아이템과 비슿한 컨텐츠를 가진 다른아이템 추천해주는방식
* '비슷함'의 기준은 장르와 키워드로 분석할수 있음
* 장르와 키워드의 값은 list안의 dict이 포함된 구조임
* string으로 변환해주어야 하기떄문에 ast패키지안에있는 literal_eval로 변환시킨다.

```python
data[['genres','keywords']].head()
```

```python
import ast as ast
```

```python
data['genres']=data['genres'].apply(ast.literal_eval)
data['keywords']=data['keywords'].apply(ast.literal_eval)
```

```python
data[['genres','keywords']].head()
```

* genres안의 값들을 dict형식으로 되어있어서 key를 가지고 뽑아올수있음
* 값을 string으로 변환해서 lambda식을써서 장르와 키워드값 변형시킴

```python
data['genres']=data['genres'].apply(lambda x : [d['name'] for d in x]).apply(lambda x:" ".join(x))
```

```python
data['keywords']=data['keywords'].apply(lambda x : [d['name'] for d in x]).apply(lambda x:" ".join(x))
```

```python
data.head()
```

## 콘텐츠기반 필터링 추천

* 장르 벡터화 : 유사 장르를 추천해주는 경우. 장르들은 띄어쓰기가 구분된 한 문자열로 저장이 되어있음. 이 문자열을 숫자로 바꿔서 벡터화 시켜야한다.
* scikit learn에 있는 CountVectorizer를 사용한다


```python
from sklearn.feature_extraction.text import CountVectorizer
```

```python
count_vector = CountVectorizer(ngram_range=(1,3))
```

```python
c_vector_genres = count_vector.fit_transform(data['genres'])
```

```python
c_vector_genres.shape
```

```python
c_vector_genres
```

## 유사도값 추출(코사인유사도)

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
similarity_genre = cosine_similarity(c_vector_genres, c_vector_genres).argsort()[:,::-1]
```

```python
print(similarity_genre)
```

```python
similarity_genre.shape
```

## 영화추천 함수 생성

* 영화제목 검색하면 유사한 장르의 영화를 추천해주게됨

```python
def recommend_movie_list(df,movie_title,top=30):
    # 특정 영화정보 뽑아내기
    target_movie_index= df[df['title'] == movie_title].index.values
    
    # 타겟 영화와 비슷한 코사인유사도값
    sim_index = similarity_genre[target_movie_index,:top].reshape(-1)
    # 본인제외
    sim_index = sim_index[sim_index!=target_movie_index]
    
    # 추천결과 새로운 df생성, 평균 평점으로 정렬
    result=df.iloc[sim_index].sort_values('score', ascending=False)[:10]
    
    return result
```

```python
recommend_movie_list(data, movie_title='Spectre')
```
