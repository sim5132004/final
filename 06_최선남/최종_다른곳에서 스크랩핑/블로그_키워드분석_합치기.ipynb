```python
import json
import pandas as pd

# JSON 파일 경로
json_paths = ['save_json[5].json', 'save_json[6].json', 'save_json[7].json', 'save_json[8].json']

# 데이터프레임 생성
df_all = pd.DataFrame(columns=["검색어", "데이터"])

# JSON 파일들 순회하며 데이터프레임에 추가
for json_path in json_paths:
    with open(json_path, "r") as json_file:
        loaded_data = json.load(json_file)
    
    # JSON 데이터의 키와 값 추출
    keys = list(loaded_data.keys())
    values = list(loaded_data.values())
    
    # 데이터프레임 생성
    df = pd.DataFrame({"검색어": keys, "데이터": values})
    
    # 현재 데이터프레임을 기존에 병합한 데이터프레임에 추가
    df_all = pd.concat([df_all, df], ignore_index=True)
```
